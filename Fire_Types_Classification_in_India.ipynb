Exploratory Data Analysis (EDA)


```python
# Check unique and n unique for all categorical features
for col in df.columns:
  if df[col].dtype == 'object':
    print(f"Column: {col}")
    print(f"Unique values: {df[col].unique()}")
    print(f"Number of unique values: {df[col].nunique()}")
    print("-" * 50)
```

```python
# Count plot for 'type'
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Distribution of Fire Types')
plt.xlabel('Fire Type')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram of 'confidence'
plt.figure(figsize=(8, 6))
sns.histplot(df['confidence'], bins=20, kde=True)
plt.title('Distribution of Confidence')
plt.xlabel('Confidence')
plt.ylabel('Frequency')
plt.show()
```

```python
# Convert 'acq_date' to datetime
df['acq_date'] = pd.to_datetime(df['acq_date'])

# Extract year from 'acq_date'
df['year'] = df['acq_date'].dt.year

# Count plot of fire types for each year
plt.figure(figsize=(12, 6))
sns.countplot(x='type', hue='year', data=df)
plt.title('Distribution of Fire Types by Year')
plt.xlabel('Fire Type')
plt.ylabel('Count')
plt.legend(title='Year')
plt.show()
```

```python
# Group by year and count fire incidents
fire_counts_by_year = df.groupby('year').size().reset_index(name='count')

# Bar plot of fire incidents by year
plt.figure(figsize=(8, 6))
sns.barplot(x='year', y='count', data=fire_counts_by_year)
plt.title('Total Fire Incidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Fire Incidents')
plt.show()
```
